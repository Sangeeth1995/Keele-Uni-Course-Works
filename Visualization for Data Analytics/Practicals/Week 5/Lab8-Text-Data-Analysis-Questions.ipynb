## Q6. Find the top 10 words using tf-idf? (rank words using idf values)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Review Text'])
print(vectorizer.get_feature_names())
print(X.shape)
print('--------------')
indices = np.argsort(vectorizer.idf_)[::-1]
features = vectorizer.get_feature_names()
top_n = 10
top_features = [features[i] for i in indices[:top_n]]
score_of_top_feature =  [vectorizer.idf_[i] for i in indices[:top_n]]
print ('Top 10 Feature=',top_features)
print('Tf-idf score=',score_of_top_feature)
```

## Q7. Apply LDA on the number the 'review text' column and find various topics. As the number of topics are not known a prior, you may have to try various values for exmaple, start with 50 topic and go till 500 topics, so on..

```python
import gensim.corpora as corpora
import gensim
from pprint import pprint
from gensim.models.ldamulticore import LdaMulticore

# pip install --upgrade gensim if you dont have gensim already installed

data= df['Review Text']
# Create doc to Bag of words becuase that is expected in Gnesim
token_data = []
for review in data:
        tokens =  review.split()
        token_data.append(tokens)
print(token_data)
print('------------')

# Create Dictionary, list of unique words
dictionary = corpora.Dictionary(token_data)
print(dictionary)
print('------------')

## create term frequency 
term_freq_corpus = [dictionary.doc2bow(text) for text in token_data]
print(term_freq_corpus)
print('------------')

# number of topics  ## Note: you can change the numer topics to what you want
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=term_freq_corpus,
                                       id2word=dictionary,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
print('------------')
doc_lda = lda_model[term_freq_corpus]
print(doc_lda)
```
